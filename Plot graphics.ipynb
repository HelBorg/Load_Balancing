```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly
import os
from subprocess import PIPE, run
import kaleido
```

```python
layout = go.Layout(plot_bgcolor='rgba(0,0,0,0)')
def add_error(df, column_name, num):
    df["Mean"] = df.mean(axis=1)
    for i in range(num):
        df[f"{column_name}_{i}"] = np.sqrt(np.power(df[f"{column_name}_{i}"] - df[f"Mean"], 2))
    df["Error"] = sum([df[f"{column_name}_{i}"] for i in range(num)])
    return df

def add_error_2(df, column_name, num):
    df["Error"] = sum([df[f"{column_name}_{i}"] for i in range(num)])
    return df

def add_error_3(df, column_name, num):
    df["Error"] = np.sqrt(sum([np.power(df[f"{column_name}_{i}"], 2) for i in range(num)]))
    return df

def add_error_4(df, column_name, num):
    df["Error"] = sum([df[f"{column_name}_{i}"] for i in range(num)])/num
    return df


def out(command):
    result = run(command, stdout=PIPE, stderr=PIPE, universal_newlines=True, shell=True)
    return result
```

For number of iterations iter run lvp and alvp.


alvp parameters specified in variable parameters and saved in file.

```python
iter = 10
for i in range(iter):
    for j in range(1):
        print(f"{i} out of {ooop}")
#         mu = 0.8 + 0.1*i
        parameters = {
            "L": 7.1,
            "mu": 0.9,
            "h": 0.2,
            "eta": 0.8,
            "gamma": [[0.07, 0.09, 0.11][0]],
            "alpha": [0.07, 0.09, 0.11][1]
        }
        np.save("cache/params.npy", parameters)
        print(parameters)

        res = os.system("start.sh")

        n = np.load("cache/agents.npy")

        layout = go.Layout(plot_bgcolor='rgba(0,0,0,0)')
        fig = go.Figure(layout=layout)
        data_lvp = {}
        for k in range(n):
            sequence = np.load(f"cache/agent_{k}_sequence_lvp.npy")[:,0]
            data_lvp[f"Agent_{k}"] = sequence
            fig.add_trace(go.Scatter(y=sequence,
                                         x=list(range(len(sequence))),
                                         name=f"Agent {k}",
                                         mode="lines"))
            
            fig.update_layout(
                xaxis_title="t",
                yaxis_title="Queue length",
                font = dict(size = 20)
            )
            fig.update_xaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')
            fig.update_yaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')
            
        fig.show()
#         fig.write_html(f"trash/trash{ind1}.html");
        
#         ind1 = ind1 + 1
#         print(f"Index: {ind1}")
        df_lvp = add_error(pd.DataFrame(data_lvp), "Agent", n)

        fig = go.Figure(layout=layout)
        data_alvp = {}
        try:
            for k in range(n):
                sequence = np.load(f"cache/agent_{k}_sequence_alvp.npy")[:,0]
                data_alvp[f"Agent_{k}"] = sequence
                fig.add_trace(go.Scatter(y=sequence,
                                             x=list(range(len(sequence))),
                                             name=f"Agent {k}",
                                 mode="lines"))
                fig.update_layout(
                    xaxis_title="t",
                    yaxis_title="Queue length",
                    font = dict(size = 20)
                )
                fig.update_xaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')
                fig.update_yaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')
            fig.show()
            
#             fig.write_html(f"trash/trash{ind1}.html");
#             ind1 = ind1 + 1
#             print(f"Index: {ind1}")
            df_alvp = add_error(pd.DataFrame(data_alvp), "Agent", n)
            data_alvp = None
        except Exception:
            print("Exception")
            continue

        layout = go.Layout(plot_bgcolor='rgba(0,0,0,0)')
        fig = go.Figure(layout=layout)
        fig.add_trace(go.Scatter(y=df_lvp["Error"],
                                 x=df_lvp.index,
                                 name=f"LVP",
                                 mode="lines",
                        line={'dash': 'dot', 'color': 'red'}))


        fig.add_trace(go.Scatter(y=df_alvp["Error"],
                             x=df_alvp.index,
                             name=f"ALVP",
                             mode="lines",
                        line={'dash': 'solid', 'color': 'green'}))

        fig.update_layout(
            xaxis_title="t",
            yaxis_title="Error",
            legend = dict(
                yanchor="top",
                y=0.99,
                xanchor="left",
                x=0.9
            ),
            font = dict(size = 20)
        )
        fig.update_xaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')
        fig.update_yaxes(showline=True, linewidth=2, linecolor='black', gridcolor='white')

        fig.show()
#         fig.write_html(f"trash/trash{ind1}.html");
#         ind1 = ind1 + 1
#         print(f"Index: {ind1}")
        if i > 0:
            alvp_er[f"Error_{i}"] = df_alvp.loc[:,["Error"]]
            lvp_er[f"Error_{i}"] = df_lvp.loc[:, ["Error"]]
        else:
            alvp_er = df_alvp.loc[:,["Error"]].rename(columns={"Error": "Error_0"})
            lvp_er = df_lvp.loc[:, ["Error"]].rename(columns={"Error": "Error_0"})
            
    
print("Finish")
```
